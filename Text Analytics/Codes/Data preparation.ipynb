```python
import csv
file = 'drive/MyDrive/Text Analytics/Sentences_50Agree.txt'
delimiter = '@'
x_train = list()
y_train = list()
with open(file, encoding = "ISO-8859-1", newline='') as infile:
    reader = csv.reader(infile, delimiter=delimiter)
    for row in reader:
        if row:
            x_train.append(row[0])
            y_train.append(row[1])

x_test = list()
y_test = list()
with open(file, encoding = "ISO-8859-1", newline='') as infile:
    reader = csv.reader(infile, delimiter=delimiter)
    for row in reader:
        if row:
            x_test.append(row[0])
            y_test.append(row[1])
```
