```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
# prendere il 30% del file tenendo bilanciate le percentuali di Sentiment
test_data = pd.read_csv('drive/MyDrive/Text Analytics/Sentences_Merge.csv')

test_data
```

```python
def remove_punctuation():
    def remove_punctuation_from_text(input_text):
        output_list=[word for word in input_text.split() if word.isalpha()]
        return ' '.join(output_list)    
    test_data['News']=test_data['News'].apply(remove_punctuation_from_text)
def correct_words():
    def correct_text(input_text):
        list_1=[str(TextBlob(word).correct()) for word in input_text.split()]
        output_text= ' '.join(list_1)
        return output_text
    test_data['News']=test_data['News'].apply(correct_text)
```
