```python
text='drive/MyDrive/Text Analytics/training_set.csv'
with open(text,mode='r',encoding='utf-8') as inputfile:
    text = inputfile.read()
```

```python
# skipping meta info and index at the beginning of the book
text = text[7285:]
print(text[:400])
```

```python
# removing CHAPTER heading and new lines
text = text.replace('CHAPTER','')
text = text.replace('\n',' ')
text = text.replace('\r',' ')
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk import word_tokenize
tokens = word_tokenize(text)
```

```python
len(tokens)
```

```python
print(tokens[2000:2050])
```

```python
from nltk import sent_tokenize
```

```python
sentences = sent_tokenize(text)
```

```python
tok_sents = [word_tokenize(sentence) for sentence in sentences]
```
