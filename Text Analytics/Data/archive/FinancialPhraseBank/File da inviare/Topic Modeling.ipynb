```python
review_file = 'drive/MyDrive/Text Analytics/Dati/archive/all-data.txt'
```

```python
import csv
train_reviews = list()
train_sentiments = list()
with open(review_file, encoding= "ISO-8859-1", newline='') as infile:
    reader = csv.reader(infile)
    for row in reader:
        if row:
            train_sentiments.append(row[0])
            train_reviews.append(row[1])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

en_stopwords = stopwords.words('english')
positive_reviews = [review for review, sentiment in zip(train_reviews, train_sentiments) if sentiment == 'positive']
tf_vectorizer = CountVectorizer(stop_words=en_stopwords, max_df=0.5, min_df=5,max_features = 1000, ngram_range=(1,2))
ptvf_features = tf_vectorizer.fit_transform(positive_reviews)
negative_reviews = [review for review, sentiment in zip(train_reviews, train_sentiments) if sentiment == 'negative']
nv_vectorizer = CountVectorizer(stop_words=en_stopwords, max_df=0.5, min_df=5,max_features = 1000, ngram_range=(1,2))
ntvf_features = nv_vectorizer.fit_transform(negative_reviews)
neutral_reviews = [review for review, sentiment in zip(train_reviews, train_sentiments) if sentiment == 'neutral']
nf_vectorizer = CountVectorizer(stop_words=en_stopwords, max_df=0.5, min_df=5,max_features = 1000, ngram_range=(1,2))
nvf_features = nf_vectorizer.fit_transform(neutral_reviews)
print(ptvf_features.shape, ntvf_features.shape)
```

# Display and visualize topics for positive reviews

```python
from sklearn.decomposition import LatentDirichletAllocation
# build topic model on positive sentiment review features
n_components = 10

lda = LatentDirichletAllocation(n_components=n_components, max_iter=10,
                                learning_method = 'batch',
                                n_jobs=-1,verbose=1)
lda.fit(ptvf_features)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        message = f'Topic {topic_idx}: '
        message += ', '.join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()
```

```python
import numpy as np
pos_feature_names = np.array(tf_vectorizer.get_feature_names())
pos_weights = lda.components_
# extract and display topics and their components
pos_feature_names = np.array(tf_vectorizer.get_feature_names())
feature_idxs = np.argsort(-pos_weights)[:, :15]
topics = [pos_feature_names[idx] for idx in feature_idxs]
for idx, topic in enumerate(topics):
    print('Topic #'+str(idx+1)+':')
    print(', '.join(topic))
    print()
```

```python
import pyLDAvis
import pyLDAvis.sklearn
from sklearn.decomposition import NMF

pyLDAvis.enable_notebook()
total_topics = 10
```

```python
pyLDAvis.sklearn.prepare(lda, ptvf_features, tf_vectorizer, mds='mmds')
```

```python
pyLDAvis.sklearn.prepare(neg_lda, ntvf_features, nv_vectorizer, mds='mmds')
```
